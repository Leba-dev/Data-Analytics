```python
sns.scatterplot(data=products,
            x='price',
            y='weight',
            sizes=(20, 200))
plt.xlabel('Precio', fontsize=12)
plt.ylabel('Peso', fontsize=12)
plt.title("Relación precio/peso")
plt.show()
```

Interpretación del gráfico

No hay una relación clara entre el peso y el precio: Los puntos en el gráfico están dispersos sin formar una línea o curva definida, lo que indica que al aumentar el peso, no necesariamente aumenta o disminuye el precio.
Existe un rango de peso sin productos: Llama la atención que no hay productos con un peso entre 1.5 y 2.0 kilogramos, independientemente de su precio. Esto sugiere que podría haber algún factor o restricción que impida que se ofrezcan productos en ese rango de peso.

En resumen
Este gráfico nos muestra que el precio de un producto no depende necesariamente de su peso, y que existe una brecha en la oferta de productos con un peso específico.

```python
# Ex4 Una variable categórica y una numérica.
#realizamos un método merge ,es como un join en mysql para relacionar tablas entre sí.
df_combi = pd.merge(companies, transactions, left_on='company_id', right_on='business_id') 
plt.figure(figsize=(12, 6))                                                                   
sns.boxplot(x='country', y='amount', data=df_combi)                                       
plt.title('Transacciones por país', fontsize=16)                                   
plt.xlabel('País', fontsize=12)
plt.ylabel('Transacción (€)', fontsize=12)
plt.xticks(rotation=45)                                                                        
plt.tight_layout()                                                                             
plt.show()
```

Interpretación del gráfico

El gráfico muestra grandes diferencias en los montos de las transacciones entre países.

Países del norte vs. sur: 
Los países del norte de Europa (Alemania,Reino Unido,Noruega y Países Bajos) tienen, en promedio, transacciones mucho más altas que los países del sur (España, Francia e Italia). La media oscila menos en los paises del norte de Europa, con una clara evolución al alza.

Variabilidad: 
Algunos países, como España, Francia y China, muestran menos variación en los montos de sus transacciones, mientras que otros tienen una mayor dispersión.

Valor atípico: 
Francia presenta una transacción excepcionalmente alta, lo que rompe con la tendencia general de este país.

Posibles explicaciones: 
Las diferencias en los montos de las transacciones podrían deberse a factores culturales, geográficos, económicos y políticos.

Recomendaciones:
Políticas para estimular el consumo: Los países con menores transacciones (como España) podrían implementar políticas para aumentar el consumo, como estudios de mercado, identificación de productos demandados y programas de incentivos.
Mantenimiento del consumo: Los países con altas transacciones (como Estados Unidos) deberían enfocarse en estrategias para mantener el nivel de consumo actual, como bonificaciones o programas de fidelización.

En resumen: 
El gráfico revela una marcada desigualdad en los hábitos de consumo entre diferentes países. Para abordar esta situación, se proponen acciones específicas para estimular el consumo en unos casos y mantenerlo en otros.Las medianas no siguen una tendencia clara en todos los paises, si que es cierto que en europa se ve una tendencia al alza. Algun pais tiene algun dato anomalos ej Francia.
